6. What are HTTP methods used in RESTful APIs ?

ans. In RESTful APIs, HTTP methods (also called verbs) define how clients interact with resources on a server. Each method has a specific function:

a)GET – Retrieve Data
Used to fetch data from the server.

Safe (does not modify data) and idempotent (same request returns the same result).

b)POST – Create Data
Used to send data to the server to create a new resource.

Not idempotent (multiple calls create multiple entries).

c)PUT – Update a Resource (Full Update)
Used to replace an existing resource with new data.

Idempotent (sending the same request multiple times gives the same result).
d) PATCH – Update a Resource (Partial Update)
Used to modify part of a resource instead of replacing it entirely.

Not necessarily idempotent (depends on the implementation).

e)DELETE – Remove a Resource
Used to delete a resource from the server.

Idempotent (deleting an already deleted resource should return the same result).

f)OPTIONS – Retrieve Allowed Methods
Used to check which HTTP methods are allowed on a resource.

g) HEAD – Retrieve Headers Only
Similar to GET, but without the response body.

Useful for checking resource existence or metadata.

8. What is the difference between GET and POST HTTP methods?

ans. 1. GET Method (Retrieve Data)

Used for: Retrieving data from the server.

Safe & Idempotent: Does not modify data and multiple requests return the same result.

Parameters: Sent in the URL as query parameters (?key=value).

Cacheable: Can be cached by browsers and proxies.

     2.POST Method (Send Data)

 Used for: Creating new resources or sending data to the server.

Not Idempotent: Sending the same request multiple times creates multiple records.

Parameters: Sent in the request body (not visible in the URL).

Not Cacheable: Data is always processed by the server.

```python
from flask import Flask, render_template, redirect, url_for
from flask_wtf import FlaskForm
from wtforms import StringField, EmailField
from wtforms.validators import InputRequired, Email

app = Flask(__name__)
app.config['SECRET_KEY'] = 'mysecretkey'  # Needed for CSRF protection

# Create a form class using WTForms
class ContactForm(FlaskForm):
    name = StringField('Name', validators=[InputRequired()])
    email = EmailField('Email', validators=[InputRequired(), Email()])

@app.route('/', methods=['GET', 'POST'])
def home():
    form = ContactForm()
    if form.validate_on_submit():
        name = form.name.data
        email = form.email.data
        return f"Form submitted successfully! Name: {name}, Email: {email}"
    return render_template('contact_form.html', form=form)

if __name__ == '__main__':
    app.run(debug=True)
```

```python
# How can you validate form data in Flask
from flask import Flask, render_template, request, redirect, url_for

app = Flask(__name__)

@app.route('/', methods=['GET', 'POST'])
def home():
    if request.method == 'POST':
        name = request.form.get('name')
        email = request.form.get('email')

        # Basic Validation
        if not name or not email:
            return "Both name and email are required!"
        if '@' not in email:
            return "Invalid email format!"

        return f"Form submitted successfully! Name: {name}, Email: {email}"

    return render_template('form.html')

if __name__ == '__main__':
    app.run(debug=True)
```

```python
#How do you manage sessions in Flask
from flask import Flask, session, redirect, url_for, request, render_template

app = Flask(__name__)

# Set a secret key to sign the session data
app.config['SECRET_KEY'] = 'mysecretkey'

@app.route('/')
def index():
    # Check if a user is logged in (session variable exists)
    if 'username' in session:
        return f'Logged in as {session["username"]} <br> <a href="/logout">Logout</a>'
    return 'You are not logged in <br> <a href="/login">Login</a>'

@app.route('/login', methods=['GET', 'POST'])
def login():
    if request.method == 'POST':
        username = request.form['username']
        session['username'] = username  # Store the username in session
        return redirect(url_for('index'))  # Redirect to the home page

    return '''
        <form method="POST">
            Username: <input type="text" name="username">
            <input type="submit" value="Login">
        </form>
    '''

@app.route('/logout')
def logout():
    session.pop('username', None)  # Remove the username from session
    return redirect(url_for('index'))  # Redirect to the home page

if __name__ == '__main__':
    app.run(debug=True)
```

```python
#How do you handle errors in Flask (e.g., 404)?
from flask import Flask, render_template

app = Flask(__name__)

@app.route('/')
def home():
    return 'Welcome to the home page!'

@app.route('/about')
def about():
    return 'This is the about page!'

# Handling 404 error
@app.errorhandler(404)
def page_not_found(error):
    return render_template('404.html'), 404

if __name__ == '__main__':
    app.run(debug=True)
```

```python
#How do you structure a Flask app using Blueprints?
from flask import Blueprint, render_template, request, redirect, url_for, flash

# Create Blueprint
auth = Blueprint('auth', __name__)

@auth.route('/login', methods=['GET', 'POST'])
def login():
    if request.method == 'POST':
        username = request.form.get('username')
        password = request.form.get('password')
        if username == "admin" and password == "admin":  # Simple authentication logic
            flash("Login successful!", "success")
            return redirect(url_for('main.home'))
        else:
            flash("Invalid credentials. Try again.", "danger")
    return render_template('login.html')

@auth.route('/register', methods=['GET', 'POST'])
def register():
    if request.method == 'POST':
        flash("Registration successful!", "success")
        return redirect(url_for('auth.login'))
    return render_template('register.html')
```

```python
#How do you define a custom Jinja filter in Flask
from flask import Flask

def uppercase_filter(s):
    """Custom Jinja filter to convert text to uppercase."""
    return s.upper()

def create_app():
    app = Flask(__name__)

    # Register the custom filter
    app.jinja_env.filters['uppercase'] = uppercase_filter

    return app

```

```python
#How can you redirect with query parameters in Flask?
from flask import Flask, redirect, url_for, request

app = Flask(__name__)

@app.route('/')
def home():
    return '<h1>Home Page</h1><a href="/redirect-example">Go to Redirect</a>'

@app.route('/redirect-example')
def redirect_example():
    # Redirect to 'result' route with query parameters
    return redirect(url_for('result', name='Ace', age=25))

@app.route('/result')
def result():
    name = request.args.get('name')  # Get query parameters
    age = request.args.get('age')
    return f'<h1>Welcome, {name}! You are {age} years old.</h1>'

if __name__ == '__main__':
    app.run(debug=True)
```
